```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
frame = pd.read_csv(r"/content/drive/MyDrive/CarDataSet.csv")
df = pd.DataFrame(frame)
df
```

```python
df.info()
```

```python
# data cleaning process
df = df.drop('New_Price',axis=1)
df = df.drop('Unnamed: 0',axis=1)
df
```

```python
# Removing NaN values from the dataframe
df = df.dropna()
df
```

```python
# Converting the catogorical data types into proper data type for further analysis
df['Mileage'] = df['Mileage'].astype(str).str.rstrip('Km/kg''kmpl').astype(float)
df['Engine'] = df['Engine'].astype(str).str.rstrip('CC').astype(int)
df['Power'] = df['Power'].astype(str).str.rstrip('bhp')
df['Power'] = df['Power'].str.replace('null', '0')
df['Power']=df['Power'].astype(float)
```

```python
df.info()
```

```python
# Spliting the car brand names from the name column of the dataframe
df['Brand'] = df['Name'].str.split(' ', expand=True)[0]
```

```python
# Identify the brand names and convert them into a list
brand = df['Brand'].unique()
```

```python
Maruti = df[df['Brand'] == 'Maruti']
Hyundai = df[df['Brand'] == 'Hyundai']
Honda = df[df['Brand'] == 'Honda']
Audi = df[df['Brand'] == 'Audi']
Nissan  = df[df['Brand'] == 'Nissan']
Toyota  = df[df['Brand'] == 'Toyota']
Volkswagen  = df[df['Brand'] == 'Volkswagen']
Tata = df[df['Brand'] == 'Tata']
Land = df[df['Brand'] == 'Land']
Mitsubishi  = df[df['Brand'] == 'Mitsubishi']
MercedesBenz  = df[df['Brand'] == 'Mercedes-Benz']
BMW = df[df['Brand'] == 'BMW']
Mahindra  = df[df['Brand'] == 'Mahindra']
Ford = df[df['Brand'] == 'Ford']
Porsche  = df[df['Brand'] == 'Porsche']
Datsun = df[df['Brand'] == 'Datsun']
Jaguar  = df[df['Brand'] == 'Jaguar']
Volvo  = df[df['Brand'] == 'Volvo ']
Chevrolet  = df[df['Brand'] == 'Chevrolet']
Skoda = df[df['Brand'] == 'Skoda']
Mini = df[df['Brand'] == 'Mini']
Fiat = df[df['Brand'] == 'Fiat']
Jeep  = df[df['Brand'] == 'Jeep']
Smart = df[df['Brand'] == 'Smart']
Ambassador  = df[df['Brand'] == 'Ambassador']
Isuzu = df[df['Brand'] == 'Isuzu']
ISUZU = df[df['Brand'] == 'ISUZU']
Force = df[df['Brand'] == 'Force']
Bentley = df[df['Brand'] == 'Bentley']
Lamborghini  = df[df['Brand'] == 'Lamborghini']
```

```python
# Checking mileage vs price for each brand
for i in brand:
    plt.figure(figsize=(7,7))
    plt.bar(df[df['Brand'] == i]['Mileage'], df[df['Brand'] == i]['Price'],label=i,)
    plt.title('Mileage Vs. Price')
    plt.xlabel('Mileage')
    plt.ylabel('Price')
    plt.legend()
    plt.show()
```

```python
# Model building
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
x= df[['Power','Year','Mileage','Engine']].values
y= df['Price'].values.reshape(-1,1)
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=100)
regression=LinearRegression()
regression.fit(xtrain,ytrain)
```

```python
regression.intercept_
```

```python
regression.coef_
```

```python
prediction = regression.predict(xtest)
```

```python
prediction
```

```python
print("Mean Squared Error is :", mean_squared_error(ytest,prediction))
print("Mean Absolute Error is :",mean_absolute_error(ytest,prediction))
print("Root Mean Squared  Error is :",np.sqrt(mean_squared_error(ytest,prediction)))
print("R2 Score is  :", r2_score(ytest,prediction))
```

```python
data = pd.DataFrame({"Actual":ytest.squeeze(), "Predicted":prediction.squeeze()})
print(data)
```
